```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import pandas as pd
import numpy as np
import plotly.io as pio

```

```python
import plotly.express as px
```

```python
data=pd.read_csv(r'C:\Users\harik\Contacts\Desktop\python\rawdata\datasets\csv-data-frames\ibmhacktrain\train.csv')
meal=pd.read_csv(r'C:\Users\harik\Contacts\Desktop\python\rawdata\datasets\csv-data-frames\ibmhacktrain\meal_info.csv')
center=pd.read_csv(r'C:\Users\harik\Contacts\Desktop\python\rawdata\datasets\csv-data-frames\ibmhacktrain\fulfilment_info.csv')
```

```python
data=data.merge(center,on='center_id',how='left')
```

```python
data=data.merge(meal,on='meal_id',how='left')
```

```python
vizdata = data.groupby(['week','center_id','center_type']).agg({'num_orders':'sum','checkout_price':'sum'})
vizdata1=vizdata.reset_index()
vizdata1
```

```python

fig=px.scatter(vizdata1, x="center_id", y="num_orders",size="checkout_price",size_max=20, color="center_type",
           animation_frame="week" ,range_x=[1,200], range_y=[500,48000],
           labels=dict(checkout_price="CheckOut Price", center_id="Center id", num_orders="Number of Orders"))
pio.write_html(fig, file=r'C:\Users\harik\Contacts\Desktop\html charts\chart1.html', auto_open=True)

```
